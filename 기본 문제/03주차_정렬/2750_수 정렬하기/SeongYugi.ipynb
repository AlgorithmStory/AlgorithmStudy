```python
N = int(input())
arr =[]

for i in range(N) :
    num = int(input())
    arr.append(num)

#선택정렬
for j in range(len(arr)) :
    min_index = j
    for k in range(j+1,len(arr)) :
        if arr[min_index] > arr[k] :
            min_index = k 
    arr[j], arr[min_index] = arr[min_index], arr[j]

#print(arr)

for l in arr :
    print(l)
```
